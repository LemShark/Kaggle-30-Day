```python
import numpy as np
import pandas as pd 
from sklearn import model_selection
```

```python
df_train = pd.read_csv("train.csv")
df_train["cat1_A"] = df_train["cat1"].apply(lambda x: 1 if x == "A" else 0)
df_train["cat8_C"] = df_train["cat8"].apply(lambda x: 1 if x == "C" else 0)
df_train["cat8_E"] = df_train["cat8"].apply(lambda x: 1 if x == "E" else 0)
df_train = df_train.drop(['cat0', 'cat1', 'cat2', 'cat3', 'cat4', 'cat6', 'cat7', 'cat9'], axis = 1)
```

```python
df_train.head()
```

```python
df_train["kfold"] = -1
```

```python
kf = model_selection.KFold(n_splits=5, shuffle=True, random_state=42)
for fold, (train_indicies, valid_indicies) in enumerate(kf.split(X=df_train)):
    df_train.loc[valid_indicies, "kfold"] = fold
```

```python
df_train.to_csv("train_folds.csv", index=False)
```
